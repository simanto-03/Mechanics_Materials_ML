```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt

y_1 = np.array([1,2,3,4,5])
y_2 = np.array([1,4,9,16,25])

f1 = {'family':'serif', 'color':'darkred', 'size':15}
f2 = {'family':'serif', 'color':'darkgreen', 'size':20}

plt.plot(y_1, label = '1', ls = ':', color='r', lw = 2.5, marker = 'o', ms = 10, mec = 'b', mfc = 'w')
plt.plot(y_2, label = '2', ls = '--', color = 'k', lw = 2.2, marker = '*', ms = 10, mec = 'r', mfc = 'w')
plt.legend(title = 'BOTH')
plt.grid(ls = "-.", color = 'c', lw = 1.2)

plt.xlabel("x-axis", loc = 'center', fontdict = f1)
plt.ylabel("y-axis", loc='center', fontdict = f1)
plt.title("curve plot", loc = 'center', fontdict = f2)

plt.xlim(0, 5)
plt.ylim(0, 30)

plt.show()
```

```python
# Scatter Graph

import numpy as np
import matplotlib.pyplot as plt

f1 = {'family':'serif', 'color':'darkred', 'size':10}
f2 = {'family':'serif', 'color':'darkgreen', 'size':15}

x = np.random.randint(200, size = (50))
y = np.random.randint(200, size = (50))
size = 3 * np.random.normal(150,50,50)
color = np.random.uniform(20,100,50)

plt.scatter(x, y, cmap = 'viridis', c = color, s = size, alpha = 0.5)
plt.colorbar()

plt.xlabel("x-axis", loc = 'center', fontdict = f1)
plt.ylabel("y-axis", loc = 'center', fontdict = f1)
plt.title("Scatter Plot", loc = 'center', fontdict = f2)

plt.show()
```

```python
# Bar Graph

import numpy as np
import matplotlib.pyplot as plt

f1 = {'family':'serif', 'color':'darkred', 'size':10}
f2 = {'family':'serif', 'color':'darkgreen', 'size':15}

x = np.array(['DHAKA','CTG','RAJ','KHUL','SYL'])
y = np.array([251,160,130,100,265])
clr = np.array(['r','k','b','c','m'])

plt.bar(x, y, color = clr, width = .2)

plt.xlabel("x-axis", loc = 'center', fontdict = f1)
plt.ylabel("y-axis", loc = 'center', fontdict = f1)
plt.title("Bar Plot", loc = 'center', fontdict = f2)

plt.show()
```

```python
# Barh Graph

import numpy as np
import matplotlib.pyplot as plt

f1 = {'family':'serif', 'color':'darkred', 'size':10}
f2 = {'family':'serif', 'color':'darkgreen', 'size':15}

x = np.array(['DHAKA','CTG','RAJ','KHUL','SYL'])
y = np.array([251,160,130,100,265])
clr = np.array(['r','k','b','c','m'])

plt.barh(x, y, color = clr, height = .2)

plt.xlabel("y-axis", loc = 'center', fontdict = f1)
plt.ylabel("x-axis", loc = 'center', fontdict = f1)
plt.title("Barh Plot", loc = 'center', fontdict = f2)

plt.show()
```

```python
# Histogram Graph

import numpy as np
import matplotlib.pyplot as plt

f1 = {'family':'serif', 'color':'darkred', 'size':10}
f2 = {'family':'serif', 'color':'darkgreen', 'size':15}

x = np.random.normal(150,10,500)

plt.hist(x, 15, color = 'r')

plt.xlabel("x-axis", loc = 'center', fontdict = f1)
plt.ylabel("y-axis", loc = 'center', fontdict = f1)
plt.title("Histogram Plot", loc = 'center', fontdict = f2)

plt.show()
```
